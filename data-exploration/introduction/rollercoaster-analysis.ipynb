# Step 2: Data Preperation
- Dropping irrelevant columns and rows
- Identifying duplicated columns
- Renaming Columns
- Feature Creation

```python
df = df[['coaster_name',
         # 'Length', 'Speed',
         'Location', 'Status',
         # 'Opening date', 'Type',
         'Manufacturer',
         # 'Height restriction', 'Model', 'Height',
         # 'Inversions', 'Lift/launch system', 'Cost', 'Trains', 'Park section',
         # 'Duration', 'Capacity', 'G-force', 'Designer', 'Max vertical angle',
         # 'Drop', 'Soft opening date', 'Fast Lane available', 'Replaced',
         # 'Track layout', 'Fastrack available', 'Soft opening date.1',
         # 'Closing date', Opened', 'Replaced by', 'Website',
         # 'Flash Pass Available', 'Must transfer from wheelchair', 'Theme',
         # 'Single rider line available', 'Restraint Style',
         # 'Flash Pass available', 'Acceleration', 'Restraints', 'Name',
         'year_introduced',
         'latitude',
         'longitude',
         'Type_Main',
         'opening_date_clean',
         # 'speed1', 'speed2', 'speed1_value', 'speed1_unit',
         'speed_mph',
         # 'height_value', 'height_unit',
         'height_ft',
         'Inversions_clean', 'Gforce_clean']].copy()
df.head()
```

```python
df.shape
```

```python
# change opening_date_clean from object to datetime
df['opening_date_clean'] = pd.to_datetime(df['opening_date_clean'])

# Rename our columns
df = df.rename(columns={'coaster_name': 'coaster_name',
                        'Location': 'location',
                        'Status': 'status',
                        'Manufacturer': 'manufacturer',
                        'Type_Main': 'type_main',
                        'opening_date_clean': 'opening_date',
                        'Inversions_clean': 'inversions',
                        'Gforce_clean': 'gforce'})
```

```python
# check for null values
df.isna().sum()
```

```python
# check duplicated rows
df.loc[df.duplicated()]
```

```python
# Check for duplicate coaster name
df.loc[df.duplicated(subset=['coaster_name'])]
```

```python
# Checking an example duplicate and to understand the reason
df.query('coaster_name == "Crystal Beach Cyclone"')
```

```python
# Drop all duplicated columns that have the same coaster name, location and opening date
df = df.loc[~df.duplicated(subset=['coaster_name','location','opening_date' ])].reset_index(drop=True).copy()
```

```python
df.shape
```

```python
# Check if we have successfully removed any duplicates
df.loc[df.duplicated(subset=['coaster_name'])]
```

# Step 4: Feature Relationships
- Scatterplot
- Heatmap Correlation
- Pairplot
- Groupby comparisons

```python
# Finding relationships between two variables using scatterplot
df.plot(kind='scatter', x='speed_mph', y='height_ft', title='Coaster Speed vs. Height')
plt.show()
```

```python
# Using Seaborn can provide us we the same result, but with most customizability
# Looking at the code below, you can see another 3rd variable added into the scatterplot
p = sns.scatterplot(x='speed_mph', y='height_ft', data=df, hue='year_introduced')
p.set_title('Coaster Speed vs. Height')
plt.show()
```

```python
# Now, we look at how we can compare multiple features/variables in a dataset
# So, pairplot essentially compares a variable with all the other variables
sns.pairplot(data=df,vars=['year_introduced','speed_mph', 'height_ft','inversions','gforce'], hue='type_main' )
plt.show()
```

```python
# We have seen how each feature's relationship with other features using pairplot.
# But, what if we want to see how correlated they are in a mathematical perspective?
# We can use pandas.DataFrame correlation function
df_corr = df[['year_introduced','speed_mph', 'height_ft','inversions','gforce']].dropna().corr()
df_corr
```

```python
# We can display the correlations we did and display it onto a heatmap
# This provides another layer of visuals to bring a different perspective when looking at the data
sns.heatmap(data=df_corr, annot=True)
plt.show()
```

# Step 5: Get insights from the data
- Try to answer a question you have about the data using a plot or statistic.

What are the locations that have at least 10 coasters that are fast?

```python
df['location'].value_counts()
```

```python
df_question = df.query('location != "Other"').groupby(['location'])['speed_mph'].aggregate(['mean', 'count']).reset_index().query('count >= 10').sort_values(by='mean', ascending=False)
df_question = df_question.rename(columns={'mean': 'mean_speed', 'count': 'coaster_count'})
df_question
```

```python
df_question.plot(kind='barh', title='Average Coaster Speed by Location that has at least 10 coasters', x='location', y='mean_speed', xlabel='Average Speed (mph)', ylabel='Location', legend=False, figsize=(12, 8))
plt.show()
```
